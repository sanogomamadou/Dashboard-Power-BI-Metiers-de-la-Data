```python
import pandas as pd

df = pd.read_csv('job_postings_flat.csv', on_bad_lines='skip')

df
```

```python
df = df.drop_duplicates()
print(f"Number of rows after removing duplicates: {df.shape[0]}")
```

```python
num_rows_with_null_salaries = df[(df['salary_year_avg'].isnull()) & (df['salary_hour_avg'].isnull())].shape[0]
print(f"Number of rows with both salary_year_avg and salary_hour_avg empty: {num_rows_with_null_salaries}")
```

```python
df['salary_year_avg'] = pd.to_numeric(df['salary_year_avg'], errors='coerce')
df['salary_hour_avg'] = pd.to_numeric(df['salary_hour_avg'], errors='coerce')
```

```python
print(df.dtypes)
```

```python
import plotly.express as px

annual_salary_df = df.dropna(subset=['salary_year_avg'])

fig_annual = px.histogram(annual_salary_df, x='salary_year_avg', title='Distribution of Annual Salaries')
fig_annual.show()

hour_salary_df = df.dropna(subset=['salary_hour_avg'])

fig_hour = px.histogram(hour_salary_df, x='salary_hour_avg', title='Distribution of Hourly Salaries')
fig_hour.show()
```
